1. Write a Pandas program to convert a Panda module Series to Python list and
it's type.

```python
import pandas as pd
d = {'a': 1, 'b': 2, 'c': 3}
ser = pd.Series(d, index = ['a', 'b', 'c'])
print(ser.to_list())
print(ser.dtype)
```

2. Write a Pandas program to add, subtract, multiple and divide two Pandas
Series.
Sample Series: [2, 4, 6, 8, 10], [1, 3, 5, 7, 9]

```python
import pandas as pd
d1 = pd.Series([2, 4, 6, 8, 10])
d2 = pd.Series([1,3,5,7,9])
add = d1 + d2
sub = d1 - d2
mul = d1 * d2
div = d1 / d2

add
sub
mul
div
```

3. Write a Pandas program to convert Series of lists to one Series.
Original Series of list
0 [Red, Green, White]
1 [Red, Black]
2 [Yellow]

```python

import pandas as pd
original = pd.Series([['Red', 'Green', 'White'], ['Red', 'Black'], ['Yellow']])
new = original.apply(pd.Series).stack().reset_index()
new
```

4. Write a Pandas program to create the mean and standard deviation of the
data of a given Series.

```python
import pandas as pd
series = pd.Series([1,2,3], [11,22,33])
me = series.mean()
sdev = series.std()
print(me)
print(sdev)
```
